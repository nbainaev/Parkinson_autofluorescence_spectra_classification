```python
line = data_add["ОАК"][4].split(": ")[1].split(", ")
line = [x.split(" - ") for x in line]
# line.remove(line[17])
# line.remove(line[16])
line = [[x[0], [x[1].split()[0]]] for x in line]
head_oak = [x[0] for x in line]
head_oak
```

```python
import os
new_data_oak = dict([(name, []) for name in head_oak])
stop = 0
all_names_park = os.listdir(".\\raw_data\\parkinson_group")
for i in range(len(data_add)):
    name_ill = data_add["номер в базе данных"][i]
    if (name_ill + "(П).txt" in all_names_park) and (name_ill + "(Л).txt" in all_names_park):
        n = 16
    else:
        n = 8
    if not pd.isna(data_add["ОАК"][i]) and "нет" not in data_add["ОАК"][i]:
        line = data_add["ОАК"][i].split(": ")[1].split(", ")
        for name in head_oak:
            for x in line:
                if name in x:
                    new_data_oak[name] += [x.split(" - ")[1].split()[0]] * n
                    stop = 1
                    break
            if stop != 1:
                new_data_oak[name] += [np.nan] * n
            stop = 0
    else:   
        for key in new_data_oak.keys():
            new_data_oak[key] += [np.nan] * n
```
